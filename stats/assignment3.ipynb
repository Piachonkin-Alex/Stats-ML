Часто, в реальной жизни, мы сталкиваемся с данными в которых есть выбросы возникшие, например, из-за опечаток. Пусть в выборке содержится 99 точек на интервале $[0;1]$ и один элемент равный 300. Тогда среднее значение нашей выборки будет не меньше трех. Ясно, что так дело не пойдет, что же делать?

Можно использовать медиану или **усеченное среднее**.

**Усеченное среднее порядка $\alpha$** это среднее составленное из выборки от которой отрезаны $\alpha$ и $1 - \alpha$ квантили. Как ее получить?


1.   Отсортировать выборку
2.   Отрезать первые и последние ($\alpha \cdot \text{размер выборки}$) элементов выборки
3.   Посчитать среднее полученной выборки

Напишите функцию для подсчета этой статистики.

```python
import math

def truncated_mean(sample, alpha=0.3):
    sort_samp = np.sort(sample)
    return np.mean(sort_samp[int(alpha * len(sort_samp)) : len(sort_samp) - 1 - int(alpha * len(sort_samp))])
```

**Теорема**
Пусть распределение симметрично относительно нуля и имеет одну моду в нуле, тогда асимптотическая дисперсия усеченного среднего порядка $\alpha$ может быть выражена по формуле:
$$
\sigma^2(\alpha) = 2\cdot\frac{\int_0^{u_{1 - \alpha}} x^2 dF(x) + \alpha\cdot(u_{1 - \alpha})^2}{(1 - 2\alpha)^2},
$$

где $u_\alpha$ – квантиль порядка $\alpha$.

Напишите функцию, которая это делает.

*Подсказка.* Для численного интегрирования используйте функцию `quad` из модуля `scipy.integrate`

```python
from scipy.integrate import quad

def asymp_var(distr, alpha):
    integ_func = lambda key : np.power(key, 2) * distr.pdf(key)
    sigma_squared = quad(integ_func, 0, distr.ppf(1 - alpha))[0] + alpha * np.power(distr.ppf(1-alpha), 2)
    return 2 * sigma_squared / np.power(1 - 2 * alpha, 2)
```

Постройте графики для фсимптотической дисперсии в зависимости от $\alpha$ для всех трех распределений на назных осях (используйте `plt.subplots(1, 3, figsize=(21,7))`). **Не забудьте про цвета**. Для загаловка графика используйте `ax[i].set_title`, для заголовка всей картинки используйте `fig.suptitle`.

```python
alpha_grid = np.linspace(0.05, 0.49999, 100)

fig, ax = plt.subplots(1, 3, figsize=(21,7))
fig.suptitle(r'Графики зависимости асимтотической дисперсии от $\alpha$', fontsize = 25)

titles = ['Стандартное нормальное распределение',
          'Стандартное распредение Лапласа', 'Стандартное распределение Коши']
distribs = [norm_distr, laplacian_distr, cauchy_distr]
colors = [color_for_normal, color_for_laplace, color_for_cauchy]
for ind, tit, col, distr in zip(range(0 ,3),titles, colors, distribs):
    ax[ind].plot(alpha_grid, [asymp_var(distr, alpha) for alpha in alpha_grid], color=col)
    ax[ind].set_title(tit)
    ax[ind].set_xlabel(r'$\alpha$')
    ax[ind].set_ylabel(r'$\sigma^2(\alpha)$')
    

plt.show()
```

**Ниже посчитайте все оценки необходимые для всех распределений**

```python
alpha = 0.35

def calc_estimations(sample, alpha):
    mean = np.cumsum(sample) / np.arange(1, len(sample) + 1)
    median = np.array([np.median(sample[:i]) for i in np.arange(1, len(sample) + 1)])
    trunc_mean = np.array([truncated_mean(sample[:i], alpha=alpha) for i in np.arange(1, len(sample) + 1)])
    return mean, median, trunc_mean
```

```python
def make_plot(sample, title, color):
    plt.figure(figsize=(16, 9))
    
    mean, median, trunc_mean = calc_estimations(sample, alpha)
    
    plt.plot(np.arange(1, len(sample) + 1), mean, label='mean', color=color)
    
    plt.plot(np.arange(1, len(sample) + 1), median, label='median', linestyle=':', color=color)
    
    plt.plot(np.arange(1, len(sample) + 1), trunc_mean, label='truncated mean', linestyle='--', color=color)

    plt.xlabel('sample size')
    plt.ylabel('estimations values')
    plt.title(title)
    plt.legend()
    plt.show
```

Найдите ОМП для двух этих распределений в предположении, что обе выбоки из многомерного нормального распределения. Не мудрите, $\mu$ и $\Sigma$ получаются в одну команду через `numpy`. Далее получите плотности для сетки `X`, `Y`. Используйте `sps.multivariate_normal`.

```python
def calc_gauss(sample):
    mu = sample.mean(axis = 0)
    sigma = (sample - mu).T @ (sample - mu)  / len(sample)
    return mu, sigma
```

```python
mu_good_wines, sigma_good_wines = calc_gauss(good_wines[['alcohol', 'citric acid']].to_numpy())
mu_bad_wines, sigma_bad_wines = calc_gauss(bad_wines[['alcohol', 'citric acid']])
```

Нарисуйте график плотности с линиями контура и нанесите точки выборки. Используйте зеленый цвет для вина плохого качества (`get_density_cmap("Greens")` и `"g"` для точек) и синий для вина хорошего качества (`get_density_cmap("Blues")` и `"b"`). Используйте функции `plt.pcolormesh`, `plt.clabel` и `plt.scatter`. Не забудьте про `plt.xlabel`, `plt.ylabel` и `plt.legend`.

```python
from matplotlib.colors import ListedColormap

def get_density_cmap(name):
    cmap =  plt.get_cmap(name)
    cmap_col = cmap(np.arange(cmap.N))
    cmap_col[:, -1] = np.linspace(0, 1, cmap.N)**2
    return ListedColormap(cmap_col)
```

```python
plt.figure(figsize=(16, 10))

distr_good = sps.multivariate_normal(mean=mu_good_wines, cov=sigma_good_wines)
distr_bad = sps.multivariate_normal(mean=mu_bad_wines, cov=sigma_bad_wines)
X, Y = np.meshgrid(np.linspace(8, 16, 1000), np.linspace(-0.3, 1, 1000))

axis = plt.subplot()
good_circles = plt.contour(X, Y, distr_good.pdf(np.dstack((X, Y))), colors = 'white')
good_mesh = plt.pcolormesh(X, Y, distr_good.pdf(np.dstack((X, Y))), cmap=get_density_cmap('Blues'))
bad_circles = plt.contour(X, Y, distr_bad.pdf(np.dstack((X, Y))), colors = 'white')
bad_mesh = plt.pcolormesh(X, Y, distr_bad.pdf(np.dstack((X, Y))), cmap=get_density_cmap('Greens'))
axis.figure.colorbar(good_mesh)
plt.clabel(good_circles, inline=1)
axis.figure.colorbar(bad_mesh)
plt.clabel(bad_circles, inline=1)


axis.scatter(good_wines['alcohol'], good_wines['citric acid'], color='blue', label='Хорошие вина')
axis.scatter(mu_good_wines[0], mu_good_wines[1], color='yellow', label = 'среднее по хорошим винам')
axis.scatter(bad_wines['alcohol'], bad_wines['citric acid'], color='green', label='Плохие вина')
axis.scatter(mu_bad_wines[0], mu_bad_wines[1], color='red', label = 'среднее по плохим винам')

plt.xlabel('Процент алкоголя')
plt.ylabel('Процент лимонной кислоты')
plt.title('Распределение вин')
plt.legend()
plt.show()
```

# Задача 3

Рассмотрим $X_1, ..., X_n \sim Bern(\theta)$. По сетке значений $\theta \in
    [0, 1]$ с шагом 0.01 постройте график зависимости нижней оценки дисперсии
    произвольной несмещенной оценки из неравенства Рао-Крамера от $\theta$. 

```python
N_s = [1, 25, 300, 1000]
grid = np.round(np.linspace(0, 1,101), 2)
n = 1000


for size in N_s:
    plt.figure(figsize = (16, 9))
    values = np.power(10 , 3) * (grid * (1 - grid)) / size
    plt.plot(grid, values, color = 'red')
    plt.xlabel("$\theta$")
    plt.ylabel("Нижняя оценка дисперсии, умноженная на $10^3$")
    plt.title(f"n = {size}")
    plt.show()
```

Какой можно сделать вывод (напишите в комментариях)?

**Вывод:** Зависимость нижней оценки есть только от размера выборки.

 Для каждого значения $\theta$ (для той же сетки) сгенерируйте выборку размера $n=1000$ для параметра $\theta$, посчитайте эффективную оценку $\theta$ и бутстрепную оценку дисперсии (количество бутстрепных выборок равно $1000$) этой эффективной оценки $\theta$.

```python
def calc_bernoulli_effective(sample):
    return sample.mean()

def calc_bernoulli_bootstrap_disp(sample, sz):
    fixed_sample = sample[:sz]
    bootstrap_sample = np.random.choice(fixed_sample, sz * sz).reshape([sz, sz])
    means = bootstrap_sample.mean(axis = 0)
    mean_of_means = means.mean()
    return np.power(means, 2).mean() - np.power(mean_of_means, 2)
```

```python
var_values = []
estimation = []
txt = ""
for theta in grid:
    sample = sps.bernoulli(theta).rvs(n)
    est = np.round(calc_bernoulli_effective(sample), 3)
    val_disp = np.round(calc_bernoulli_bootstrap_disp(sample, n),6)
    
    estimation.append(est)
    var_values.append(val_disp)
    val_disp *= 1000
    print(f"Параметр равен {theta}: эффективная оценка равна {est}, бустрепная эффективная оценка, умноженная на 1000 равна {np.round(val_disp, 5)}")
```

 Нарисуйте график зависимости полученных бутстрепных оценок от $\theta$.

```python
plt.figure(figsize = (16, 9))
plt.plot(grid, np.array(var_values) * 1000, color = 'red', label = 'Бустрепная оценка дисперсии')
plt.plot(grid ,1000 * (grid * (1 - grid)) / n, color = 'blue', label = 'Дисперсия эффективной оценки')
plt.title("Оценки дисперсии")
plt.xlabel(r"$\theta$")
plt.ylabel("Бустрепная оценка дисперсии эффективной оценки, умноженная на $10^3$")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(grid, estimation, label = r"$\overline{X}$")
plt.plot(grid, grid, label = r'$ y = x$')
plt.fill_between(grid, estimation - np.sqrt(var_values), estimation + np.sqrt(var_values), 
                 alpha = 0.5, label = r"$\overline{X} \pm sigma(\overline{X})$")
plt.xlabel(r'$\theta$')
plt.ylabel(r'Бутсрепная оценка $\theta$')
plt.title(r'График зависимости бутстрепных оценок от $\theta$')
plt.legend()
plt.show()
```

**Вывод:** Дисперсия, полученная из бустрепных выборок эффективной оценки хорошо приближает нижнию границу из Теоремы Рао-Крамера.
