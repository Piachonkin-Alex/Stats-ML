**Ответ:**

Часто, в реальной жизни, мы сталкиваемся с данными в которых есть выбросы возникшие, например, из-за опечаток. Пусть в выборке содержится 99 точек на интервале $[0;1]$ и один элемент равный 300. Тогда среднее значение нашей выборки будет не меньше трех. Ясно, что так дело не пойдет, что же делать?

Можно использовать медиану или **усеченное среднее**.

**Усеченное среднее порядка $\alpha$** это среднее составленное из выборки от которой отрезаны $\alpha$ и $1 - \alpha$ квантили. Как ее получить?


1.   Отсортировать выборку
2.   Отрезать первые и последние ($\alpha \cdot \text{размер выборки}$) элементов выборки
3.   Посчитать среднее полученной выборки

Напишите функцию для подсчета этой статистики.

```python
import math

def truncated_mean(sample, alpha=0.3):
    sorted_sample = np.sort(sample)
    size = sorted_sample.shape[0]
    return np.mean(sorted_sample[int(size*alpha) : size - 1 - int(size*alpha)])
```

**Теорема**
Пусть распределение симметрично относительно нуля и имеет одну моду в нуле, тогда асимптотическая дисперсия усеченного среднего порядка $\alpha$ может быть выражена по формуле:
$$
\sigma^2(\alpha) = 2\cdot\frac{\int_0^{u_{1 - \alpha}} x^2 dF(x) + \alpha\cdot(u_{1 - \alpha})^2}{(1 - 2\alpha)^2},
$$

где $u_\alpha$ – квантиль порядка $\alpha$.

Напишите функцию, которая это делает.

*Подсказка.* Для численного интегрирования используйте функцию `quad` из модуля `scipy.integrate`

```python
from scipy.integrate import quad

def asymp_var(distr, alpha):
    I = lambda x : x**2 * distr.pdf(x)
    sigma_squared = np.array([2*quad(I, 0, distr.ppf(1-a))[0] + a * (distr.ppf(1-a))**2 for a in alpha])
    return sigma_squared / (1 - 2*alpha)**2
```

Постройте графики для фсимптотической дисперсии в зависимости от $\alpha$ для всех трех распределений на назных осях (используйте `plt.subplots(1, 3, figsize=(21,7))`). **Не забудьте про цвета**. Для загаловка графика используйте `ax[i].set_title`, для заголовка всей картинки используйте `fig.suptitle`.

```python
alpha_grid = np.linspace(0.05, 0.49999, 100)

fig, ax = plt.subplots(1, 3, figsize=(21,7))

ax[0].plot(alpha_grid, asymp_var(norm_distr, alpha_grid), color=color_for_normal)
ax[0].set_title('Нормальное распределение')
ax[1].plot(alpha_grid, asymp_var(laplacian_distr, alpha_grid), color=color_for_laplace)
ax[1].set_title('Распределение Лапласа')
ax[2].plot(alpha_grid, asymp_var(cauchy_distr, alpha_grid), color=color_for_cauchy)
ax[2].set_title('Распределение Коши')

ax[0].set_xlabel(r'$\alpha$')
ax[1].set_xlabel(r'$\alpha$')
ax[2].set_xlabel(r'$\alpha$')

ax[0].set_ylabel(r'$\sigma^2(\alpha)$')
ax[1].set_ylabel(r'$\sigma^2(\alpha)$')
ax[2].set_ylabel(r'$\sigma^2(\alpha)$')

fig.suptitle(r'Графики зависимости асимтотической дисперсии от $\alpha$ для расзличных распределений', fontsize=20)
plt.show()
```

```python
def est(sample, split, alpha):
    tmean = np.array([truncated_mean(sample[:i], alpha=alpha) for i in split])
    median = np.array([np.median(sample[:i]) for i in split])
    mean = np.array([np.mean(sample[:i]) for i in split])
    return (tmean, median, mean)
```

```python
def plot_est(tmean, median, mean, split, color, title):
    plt.figure(figsize=(12, 8))

    plt.plot(split, tmean, 
             label='truncated mean', 
             linestyle='--', 
             color=color)

    plt.plot(split, median, 
             label='median', 
             linestyle=':', 
             color=color)

    plt.plot(split, mean, 
             label='mean',  
             color=color)

    plt.xlabel('sample size')
    plt.ylabel('value')
    plt.title(title)
    plt.legend()
    plt.show
```

# Задача 3

Рассмотрим $X_1, ..., X_n \sim Bern(\theta)$. По сетке значений $\theta \in
    [0, 1]$ с шагом 0.01 постройте график зависимости нижней оценки дисперсии
    произвольной несмещенной оценки из неравенства Рао-Крамера от $\theta$. 

$X_1, ..., X_n \sim Bern(\theta)$
$$
f(X) = \theta^{\sum X_i} (1 - \theta)^{n - \sum X_i}
$$
$$
u_{\theta} (X) = \frac{\partial}{\partial\theta} L_{\theta}(X) = \frac{\partial}{\partial\theta} \left(\sum\limits_{i=1}^n X_i \ln\theta + \left(n - \sum\limits_{i=1}^n X_i\right)\ln(1-\theta)\right) = \frac{n}{\theta(\theta - 1)} (\theta - \overline{X})
$$

$$
\overline{X} - \theta = \frac{\theta(1 - \theta)}{n} \cdot u_{\theta}(X)
$$
По критерию эффективности оценка $\theta^* = \overline{X}$ является эффективной оценкой $\theta$ с информацией Фишера $i(\theta) = \frac{1}{\theta(1 - \theta)}$.


Нижняя оценка для дисперсии $\frac{1}{ni(\theta)} = \frac{\theta(1 - \theta)}{n}$. Ниже мы не будем учитывать $n$ в построении графика, так как это некоторая константа для всех $\theta$.

```python
grid = np.arange(0, 1, 0.01)

lower_bound_var =  grid * (1 - grid)
```

```python
plt.figure(figsize=(12,8))

plt.plot(grid, lower_bound_var)

plt.title(r'График зависимости нижней оценки дисперсии от $\theta$')
plt.ylabel(r'$\frac{1}{i(\theta)}$')
plt.xlabel(r'$\theta$')
plt.show()
```

 Для каждого значения $\theta$ (для той же сетки) сгенерируйте выборку размера $n=1000$ для параметра $\theta$, посчитайте эффективную оценку $\theta$ и бутстрепную оценку дисперсии (количество бутстрепных выборок равно $1000$) этой эффективной оценки $\theta$.

```python
def bootstrap_mean_var(sample, bs_size):
    bootstrap = np.random.choice(sample, bs_size*bs_size).reshape(bs_size, bs_size)
    mean = np.mean(np.mean(bootstrap, axis = 0))
    var = np.mean(np.mean(bootstrap**2, axis=0) - np.mean(bootstrap, axis=0)**2)
    return [mean, var]
```

```python
n = 1000
bern_sample = np.array([sps.bernoulli.rvs(size=n, p=theta) for theta in grid])
mean_var = np.array([bootstrap_mean_var(bern_sample[i], n) for i in range(bern_sample.shape[0])]).T
```
