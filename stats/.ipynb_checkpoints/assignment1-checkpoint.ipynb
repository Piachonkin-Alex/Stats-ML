### Задача 1

Сгенерируйте выборку $X_1, \dots, X_N$ из равномерного распределения на отрезке $[0, \theta]$ для $N = 10^4$.

```python
N = 1e4  # use this
N = int(N)
theta = 1
sample = sps.uniform(loc=0, scale=theta).rvs(N)
```

Для всех $n \leqslant N$ посчитайте оценки параметра $\theta$ из теоретической задачи: $2\overline{X}, \; \overline{X} + X_{(n)}/ 2, \; (n+1)X_{(1)}, \; X_{(1)}+X_{(n)}, \; \frac{n+1}{n} X_{(n)}$. Используйте векторные операции.

__Подсказка:__ Могут быть полезными функции `np.arange`, `np.cumsum`, `np.maximum.accumulate` и `np.minimum.accumulate`

```python
def double_sample_mean_estimation(sample: np.array, theta: int) -> np.array:
    cumsums = np.cumsum(sample)
    return np.abs(theta - 2 * cumsums / (np.arange(len(sample)) + 1))

def sample_plus_max_estimation(sample: np.array, theta: int) -> np.array:
    maximums = np.maximum.accumulate(sample)
    cumsums = np.cumsum(sample)
    return np.abs(theta - (cumsums / (np.arange(len(sample)) + 1) + maximums / 2))

def min_order_statistics_estimation(sample: np.array, theta: int) -> np.array:
    minimums = np.minimum.accumulate(sample)
    return np.abs(theta - (np.arange(len(sample)) + 2) * minimums)

def min_order_plus_max_order_estimation(sample: np.array, theta: int) -> np.array:
    maximums = np.maximum.accumulate(sample)
    minimums = np.minimum.accumulate(sample)
    return np.abs(theta - (maximums + minimums))

def max_order_statistics_estimation(sample: np.array, theta: int) -> np.array:
    maximums = np.maximum.accumulate(sample)
    return np.abs(theta - (np.arange(len(sample)) + 2) / (np.arange(len(sample)) + 1) * maximums)
```

Постройте на одном графике разными цветами для всех оценок функции модуля разности оценки и истинного значения $\theta$ в зависимости от $n$. Если некоторые оценки (при фиксированном значении $n$) сильно отличаются от истинного значения параметра $\theta$, то исключите их и постройте еще один график со всеми кривыми (для измененного значения $\theta$). Для избавления от больших значений разности в начале ограничьте масштаб графика. Для наглядности точки лучше соединить линиями.

Не забудьте подписать оси, а также добавить легенду к графику.

```python
estimators = [
    double_sample_mean_estimation,
    sample_plus_max_estimation,
    min_order_statistics_estimation,
    min_order_plus_max_order_estimation,
    max_order_statistics_estimation
]

labels = ["$2\overline{X}$", "$\overline{X} + X_{(n)}/2$",
          "$(n+1)X_{(1)}$", "$X_{(1)}+X_{(n)}$", "$\dfrac{n+1}{n}X_{(n)}$"]
colors = ['red', 'black', 'yellow', 'green', 'blue']

def draw_plot(sample: np.array, theta: int) -> None:
    plt.figure(figsize = (16, 9))

    for estimator, label, color in zip(estimators, labels, colors):
        plt.plot(np.arange(1 ,N + 1),
             estimator(sample, theta),
            color = color,
            alpha = 0.4,
            linewidth = 3,
            label = label)

    plt.title(fr'Сравнение оценок для $\theta$={theta}', fontsize=25)
    plt.xlabel('Размер выборки', fontsize=20)
    plt.ylabel('Разность между параметром и оценкой', fontsize=20)
    plt.ylim(-0.001, 0.005)
    plt.xlim(300, N + 1)

    plt.legend()

    plt.show()

draw_plot(sample, theta)
```

```python
plt.figure(figsize = (16, 9))
plt.plot(np.arange(1 ,N + 1),
            min_order_statistics_estimation(sample, theta),
            color = 'purple',
            alpha = 0.4,
            linewidth = 2,
        )
plt.title(fr'Оценка $(n+1)X_{(1)}$ для $\theta$={theta}', fontsize=25)
plt.xlabel('Размер выборки', fontsize=20)
plt.ylabel('Разность между параметром и оценкой', fontsize=20)
plt.ylim(-0.001, 3)
plt.xlim(300, N + 1)


plt.show()
```

__Ответ:__ Наиболее удачной получилась оценка $\dfrac{n+1}{n}X_{(n)}$.



Проведите эксперимент для разных значений $\theta$ (количество графиков равно количеству значений $\theta$)

```python
thetas = np.round(np.linspace(0.4, 2, 5), 1)

for theta in thetas:
    sample = sps.uniform(loc=0, scale=np.round(theta, 1)).rvs(N)
    draw_plot(sample, theta)
```

### Задача 2

Придумайте распределение, у которого конечны первые два момента, а третий – нет. Сгенерируйте выборку $X_1, \dots, X_N$ из этого распределения для $N = 10^4$. 

Будем искать такое распределение в множестве абсолютно-непрерывных распределений. Пусть f(x) -- плотность искомого распределения. Тогда, необходимо, чтобы $\int_{-\infty}^{+\infty}xf(x)$ и $\int_{-\infty}^{+\infty}x^2f(x)$ -- сходились, а $\int_{-\infty}^{+\infty}x^3f(x)$ -- расходился. Воспользуемся эталонными интегралами. Мы знаем, что $\int_{1}^{+\infty}\frac{1}{x^\alpha}$ сходится при $\alpha > 1$, расходится в противном случае. Следовательно, при $f(x) = \frac{1}{x^4}$ интегралы $\int_{1}^{+\infty}xf(x)$ и $\int_{1}^{+\infty}x^2f(x)$ сходятся, а $\int_{1}^{+\infty}x^3f(x)$ -- расходится. Всего, чего осталось добиться, это чтобы $\int_{-\infty}^{+\infty}f(x) = 1$. Воспользуемся индикатором $I(x > 1)$ и значением данного эталонного интеграла, и получим $f(x) = \frac{3}{x^4}\cdot I(x > 1)$ -- распределение Паретто с параметром $k = 3$, $x_m = 1$.

```python
N = 10 ** 4
sample = sps.pareto(3).rvs(N)
```

Постройте график плотности, а также нанесите точки выборки на график (с нулевой y-координатой)

**Подсказка**: Может быть полезен параметр `alpha` в функции `plt.plot`

```python
def draw_density_plot(distribution, sample, title_label, x_lim_left, x_lim_right, density_points):
    plt.figure(figsize=(16, 9))
    zeros = np.zeros(len(sample))
    plt.plot(density_points, distribution.pdf(density_points), color='blue')
    plt.title(title_label, fontsize=20)
    plt.xlabel("x", fontsize=20)
    plt.ylabel("y", fontsize=20)
    plt.xlim(x_lim_left, x_lim_right)
    
    plt.scatter(sample, zeros, s = 69, alpha=0.01, color="green")
    
    plt.show()

    
```

```python
draw_density_plot(sps.pareto(3), sample,"Плотность распределения Парето", 0, 7, np.linspace(0, 7, 1000) )
```

Для всех $n \leq N$ посчитайте оценку $s^2 = s^2(X_1,\dots, X_N)$ для дисперсии.

```python
def calc_cum_variances(sample: np.array) -> np.array:
    cumsums = np.cumsum(sample)
    square_cumsums = np.cumsum(np.power(sample, 2))
    answer = square_cumsums / np.arange(1, len(sample) + 1) - (cumsums / np.arange(1,len(sample) + 1)) ** 2
    return answer
```

Постройте график зависимости модуля разности оценки дисперсии и ее истинного значения от $n$.

```python
def make_plot(sample, theta):
    plt.figure(figsize = (16, 9))
    plt.plot(
        np.arange(1, len(sample) + 1),
        np.abs(calc_cum_variances(sample) - theta),
        color = 'red'
    )
    plt.title('Зависимость модуля разности оценки дисперсии и ее истинного значения от $n$', fontsize=20)
    plt.xlabel('Размер выборки', fontsize=24)
    plt.ylim(0, 1)
    plt.ylabel('Значения модуля разности', fontsize=24)
    plt.show()
```

```python
theta = np.round(sps.pareto(3).std() ** 2, 3)
make_plot(sample, theta)
```

**Вывод**:
Мы знаем, что асимптотически выборочная дисперсия стремится к значению дисперсии. Однако, можно заметить, что на графике мы стремления не наблюдаем. Тем не менее, если взять более большую выборку, сходимость модуля разности к нулю будет. Такой разброс скорее всего связан с тем, что у распределения Парето нету 3-го момента, а значит, скорее всего выбранная оценка не является асимптотически нормальной

**Примечание.** Мы точно знаем, что если существует 4-ый момент, то выборочная дисперсия является асимптотически нормальной оценкой.

### Задача 3

Сгенерируйте выборку $X_1,\dots, X_N$ из стандартного нормального распределения для $N = 10^4$.

```python
sample = sps.norm().rvs(N)
```

Для всех $n \leqslant N$ посчитайте по ней эмпирическую функцию распределения.


```python
from statsmodels.distributions.empirical_distribution import ECDF  # can be useful, but not necessary

list_of_empirical_functions = [ECDF(sample[:sz]) for sz in np.arange(1, N  + 1)]
```

Для некоторых n (например, $n \in \{10,25,50,100,1000,N\}$ постройте графики эмпирической функции распределения (отметьте на оси абсцисс точки "скачков" кривых, нанеся каждую из "подвыборок" на ось абсцисс на каждом соответствующем графике с коэффициентом прозрачности `alpha=0.2`), нанеся на каждый из них истинную функцию распределения (количество графиков равно количеству различныз значений $n$).

```python
def plot_empirical_function(list_of_emprical_functions, n, grid, title_label):
    plt.figure(figsize = (16, 9))
    plt.plot(grid, 
            list_of_emprical_functions[n - 1](grid),
             color = 'red',
             label = 'Эмпирическая функция'
            )
    plt.plot(grid,
            sps.norm().cdf(grid),
            color = 'blue',
            label = 'Истинная функция распределения'
            )
    plt.title(title_label + fr' при n = {n}')
    plt.scatter(x = sample[:n], y = np.zeros(n), alpha = 0.2, label = 'Значения выборки', color = 'blue')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()
    
```

```python
n_list = [10, 25, 50, 100, 1000, 10**4]
grid = np.linspace(-5,5, 10**4)
for n in n_list:
    plot_empirical_function(list_of_empirical_functions, n, grid, "Функция распределения")
```

Для всех $n ≤ N$ посчитайте точное значение $D_n = \displaystyle{\sup_{x∈R}} | \hat{F}_n(x)−F(x)|$ и постройте график зависимости статистик $D_n$ и $\sqrt{n}D_n$ от $n$. 


```python
def plot_maximum_std_and_sqrt_std(emprical_functions, distribution, grid, N):
    maximum_std = [max(emprical_functions[sz](grid)  - distribution.cdf(grid)) for sz in np.arange(0, N)]
    maximum_std_cdot_n = np.sqrt(np.arange(1, n + 1)) * maximum_std
    values = [maximum_std, maximum_std_cdot_n]
    titles = [r"График зависимости $D_{n}$ от $n$", r"График зависимости $\sqrt{n}D_n$ от $n$"]
    for points, title in zip(values, titles):
        plt.figure(figsize = (16, 9))
        plt.title(title)
        plt.plot(grid, points, color = 'red')
        plt.show()
        
```

```python
plot_maximum_std_and_sqrt_std(list_of_empirical_functions, sps.norm, grid, N)
```

**Выводы**:
График зависимости $D_n$ говорит нам о том, что  эмпирическая функция распределения сходится равномерно к исходной функции распределения. График зависимости $\sqrt n D_n$  подтверждает, что верна теорема Колмогорова об эмпирической функции распределения. Из данного графика мы не можем понять вид распределения, но точно понимаем, что оно не является вырожденным.

### Задача 4

Сгенерируйте $N_{\text{samples}}=400$ выборок из равномерного распределения $U_{[0, \theta]} \ \theta=1$ размера $N=40$. Для каждой выборки посчитайте статистики $\hat\theta = 2\overline{X}, \ \theta^* = \frac{n+1}{n} X_{(n)}$. Постройте гистограмму получившихся значений каждой из статистик на одном графике, в качестве параметра `bins` функции `plt.hist` передайте значение ниже, а таккже передайте параметр `alpha=0.6`.

```python
N_samples = 400
N = 40

bins = [i / 40 + 0.8 for i in range(18)]

samples = sps.uniform().rvs((N_samples, N))

sample_mean_estimation = samples.mean(axis = 1) * 2
maximum_order_statistics_estimation =  samples.max(axis = 1) * (N + 1)/N
plt.figure(figsize = (10, 6))

plt.hist(
    x = sample_mean_estimation, 
    bins = bins, 
    alpha = 0.6,
    color = 'red',
    label =  r'$\hat\theta = 2\overline{X}$'
)
plt.hist(
    x = maximum_order_statistics_estimation,
    bins = bins,
    alpha = 0.6,
    color = 'green',
    label = r'$\theta^* = \frac{n+1}{n} X_{(n)}$'
)
plt.title('Гистограмма значения статистик')
plt.legend()
plt.show()
```

Постройте гистограммы для статистик $\sqrt{n} (\hat\theta - \theta)$ и $1 - n  (\theta^* - \theta)$

```python
sqrt_sample_mean_est = np.sqrt(N) * (sample_mean_estimation - 1)
linear_max_order_est = 1 - N * (maximum_order_statistics_estimation - 1)

plt.figure(figsize = (10, 6))

plt.hist(
    x = sqrt_sample_mean_est, 
    alpha = 0.6,
    density = True,
    color = 'red',
    label =  r'$\sqrt{n} (\hat\theta - \theta)$'
)
plt.hist(
    x = linear_max_order_est,
    alpha = 0.6,
    density = True,
    color = 'green',
    label = r'$1 - n  (\theta^* - \theta)$'
)


expon_grid = np.linspace(0, 5, 100)
normal_grid = np.linspace(-2, 2, 100)

plt.plot(expon_grid, sps.expon().pdf(expon_grid), label = r'$Exp(1)$')
plt.plot(normal_grid, sps.norm(0, 1/np.sqrt(3)).pdf(normal_grid),
         label = r'$N\left(0, \frac{1}{\sqrt{3}}\right)$')

plt.title('Гистограмма значения статистик')
plt.legend()
plt.show()
```

На какие распределения похожи получившиеся гистограммы?

**Ответ:**
Гистограмма $1 - n  (\theta^* - \theta)$ похожа на экспонециальное распределение c параметром 1.

Гистограмма $\sqrt{n} (\hat\theta - \theta)$ похожа на нормальное распределение с параметрами 0 и $\frac{1}{\sqrt3}$.

Вспомните чему равен коэффициент $\sigma(\theta)$ для асимптотиически нормальной оценки $\hat\theta = 2\overline{X}$ для параметра $\theta$ равномерного распределения в формуле 
$$
\sqrt{n}\frac{\left(\hat\theta - \theta\right)}{\sqrt{\sigma(\theta)}} \xrightarrow{d} N(0, 1)
$$.

**Ответ**: Соглсно ЦПТ, $\sigma(\theta) = D(2X_1) = 4D(X_1) = 4 \cdot \frac{\theta^2}{12} = \frac{\theta^2}{3}$.

Посчитайте значения статистики
$$\sqrt{n}\frac{\left(\hat\theta - \theta\right)}{\sqrt{\sigma(\theta)}}$$
для каждой выборки. Передайте получившиеся значения в переменную `theta_norm`. И запустите ячейку снизу.

```python
import statsmodels.api as sm


theta_norm = np.sqrt(3) * sqrt_sample_mean_est
fig, ax = plt.subplots(figsize=(10, 10))
sm.qqplot(theta_norm, line='45', ax=ax)
fig.show()
```

**Ответ:** Да, такой вывод сделать можно, так как график хорошо ложится на прямую $y = x$.

Вернемся к статистике $\theta - n  (\theta^* - \theta)$. Еще раз взгляните на гистограмму, соотвутствующую этой статистике.
```python
fig, ax = plt.subplots(figsize=(15, 15))
sm.qqplot(linear_max_order_est, line='45', ax=ax, dist = sps.expon)

fig.show()
```

**Вывод**: из данного графика можно утверждать, что выборка взята из стандартного экспоненциального распределения.

Плохо ли, что оценка $\theta^*$ не асимптотически нормальна? Сделайте вывод о скорости сходимости оценок. Какая из них «выгоднее»?

**Ответ:**

```python
theta = 1
N = 300
sample = sps.uniform().rvs(N)

sample_mean_est = 2 / (np.arange(N) + 1) * sample.cumsum()
max_order_est = np.maximum.accumulate(sample) * (np.arange(N) + 2) / (np.arange(N) + 1)

plt.figure(figsize = (16, 9))
grid = np.arange(N) + 1

plt.plot(grid, sample_mean_est,  linewidth = 2, 
         color = 'red', alpha = 0.4, label = r'$\hat\theta = 2\overline{X}$')
plt.plot(grid, max_order_est,  linewidth = 2,
         color = 'green', alpha = 0.4, label = r'$\theta^* = \frac{n+1}{n} X_{(n)}$')
plt.hlines(y = 1, xmin = 1, xmax = N + 1,  linewidth = 3,
           color = 'blue', alpha = 0.2, label = r'$\theta = 1$')

plt.xlabel("Размер выборки")
plt.ylabel("Значение оценки")
plt.legend()
plt.show()
```

Как можем заметить, $\theta^*$ сходится лучше, чем $\hat\theta$.

**Вывод**:
Исходя из полученных графиков можем сделать вывод, что оценки, которые  обладают свойством асимптотической нормальности, не обязательно лучше оценок, не обладающих данным свойством. Из асимптотической нормальности следует лишь состоятельность оценки, а здесь обе оценки являются сильно состоятельными.
