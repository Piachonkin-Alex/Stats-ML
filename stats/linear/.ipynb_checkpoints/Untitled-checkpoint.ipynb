Помним, что для применения линейной регрессии желательно нормализовать данные. Для того, чтобы про это не забывать, пишем маленький пайплайн.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

def make_basic_pipeline(model):
    return Pipeline([
        ("Scaler", StandardScaler()),
        ("Model", model),
    ])
```

В Ridge и Lasso моделях у нас есть параметр регуляризации. Для него нужно построить GridSearch. Сделаем для него функцию, а также общую функцию обучения. Для подсчета качества не забываем кросс-валидацию.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.exceptions import ConvergenceWarning

def find_best_params(model, params_grid, X_train, y_train):
    searcher = GridSearchCV(model,params_grid,scoring='neg_mean_squared_error')
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", category=ConvergenceWarning)
        searcher.fit(X_train, y_train)

    print("Best parameters:", searcher.best_params_)
    model = searcher.best_estimator_
    return model
```

```python
def train_and_test_cv(model, parameters, title, X, y, need_grid = False):
    best_model = model
    if need_grid:
        best_model = find_best_params(model,parameters, X, y)
    else:
        best_model.fit(X, y)
    
    preds = best_model.predict(X)
    mse_score = metrics.mean_squared_error(preds,y)
    
    print(f'MSE {title} -- ', mse_score)
    cv_score = cross_val_score(best_model, X, y, scoring='neg_mean_squared_error', cv=4).mean()
    print(f'cv score {title} cv_score-- ', cv_score)
    return mse_score, cv_score
```

```python
model =  make_basic_pipeline(LinearRegression(fit_intercept=True))
mse_lin, cv_lin = train_and_test_cv(model, None, "Linear", X, y, False)
```

```python
parameters = {'Model__alpha': np.append(np.arange(0.2, 20.1, 0.05), np.logspace(-6, -3, 4))}
model = make_basic_pipeline(Ridge(fit_intercept=True)) 
mse_ridge, cv_ridge = train_and_test_cv(model, parameters, "Ridge", X, y, True)
```

```python
parameters = {'Model__alpha': np.append(np.arange(0.01, 3, 0.01), np.logspace(-6, -3, 4))}
model = make_basic_pipeline(Lasso(fit_intercept=True)) 
mse_lass, cv_lass = train_and_test_cv(model, parameters, "Lasso",X, y, True)
```

```python
dat = {'mse' : [mse_lin,mse_ridge,mse_lass],
       'cv': [cv_lin,cv_ridge,cv_lass]}
st = pd.DataFrame(dat, index = ['Linear', 'Ridge', 'Lasso'])
st
```

Как видим, регуляризация не помогла нам в этот раз. Возможно, это связано с тем, что данные достаточно неплохо расположены и штрафовать за большие веса параметров не нужно. Теперь, попробуем поработать с приззнаками. Будем использовать ExtraTreesRegressor + LinearRegression. Мы сделаем следующим образом -- посмотрим на feature_importances_, выберем некоторое число признаков, которые мы оставим, а потом уже пройдемся по каждому количеству признаков и посмотрим, где лучшие результаты.

```python
forest = ExtraTreesRegressor(n_estimators=100)
forest.fit(X_train, y_train)
```

```python
forest.feature_importances_
```

Кажется, стоит взять 4 признака (с threshold 0.1). Зафиксируем это число.

```python
k = X.shape[1]
print(k)
scores_mse = []
scores_cv = []
for features_count in range(1, k + 1):
    tree = RFE(estimator=ExtraTreesRegressor(n_estimators=100), n_features_to_select=features_count, step=1)
    X_rfe = tree.fit_transform(X, y)
    model = make_basic_pipeline(LinearRegression(fit_intercept=True)) 
    mse, cv = train_and_test_cv(model, None, "Linear", X_rfe, y, False)
    scores_mse.append(mse)
    scores_cv.append(cv)
```

```python
dat = {'mse' :scores_mse ,
       'cv': scores_cv}
st = pd.DataFrame(dat, index = np.arange(1, k + 1))
st
```

Как видим, при количестве признаков, равных 4 и получили хороший результат. После 5+ признаков ошибка немного начинает увеличивается.
