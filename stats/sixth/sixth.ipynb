Напишем функцию для подсчета статистики Розенблатта и с помощью моделирования найдем критические значения для $\alpha = 0.01, 0.05, 0.1$. Для нашей же задачи стандартно зафиксируем $\alpha = 0.05$.

```python
num_samples = 10000
sz_first, sz_second = 200, 250
data = np.load("hw6t2.npy").T
data.shape
```

```python
def calc_Rosenblatt_stat(x, y):
    sz_first, sz_second = x.shape[0], y.shape[0]
    divider = 1 / (sz_first + sz_second)
    multipier = sz_first * sz_second
    mod = []
    for el in x:
        mod.append(el)
    for el in y:
        mod.append(el)
    mod = np.array(mod)
    ranks = np.argsort(mod).argsort()
    first_ranks = sorted(ranks[:sz_first])
    second_ranks = sorted(ranks[sz_first:])
    z = divider * (1 / 6 + np.sum(np.power(first_ranks - np.arange(sz_first), 2)) / sz_second + 
                       np.sum(np.power(second_ranks - np.arange(sz_second), 2)) / sz_first)
    return z  - (2.0 / 3.0) * multipier * divider
```

```python
def calc_critical_values_for_Rosenblatt(num_samples, sz_first, sz_second, alphas):
    first_batch = sps.norm().rvs((num_samples, sz_first))
    second_batch = sps.norm().rvs((num_samples, sz_second))
    stats_vals = np.array([])
    for x, y in zip(first_batch, second_batch):
        stats_vals = np.append(stats_vals, calc_Rosenblatt_stat(x, y))
    stats_vals = sorted(stats_vals)
    return [stats_vals[int((1 - alpha) * len(stats_vals))] for alpha in alphas]
```

```python
critical = calc_critical_values_for_Rosenblatt(num_samples, sz_first, sz_second, [0.01, 0.05, 0.1])
critical
```

Кажется, что очень неплохо приблизились к табличным значениям, значит все сделано верно. Теперь уже взлянем на данные из нашей задачи. 

```python
plt.figure(figsize = (16, 9))
plt.hist(data)
plt.hist(data[0], color = 'red', alpha = 0.5)
plt.hist(data[1], color = 'blue', alpha = 0.5)
plt.show()
```

Кажется, что распредления все же немного отличаются. Проверим на уровне значимости 0.05.

```python
stat_val = calc_Rosenblatt_stat(data[0], data[1]) 
print(f"Значение статистики -- {stat_val}")
if stat_val < critical[1]:
    print(f"Гипотеза об однородности на уровне значимости 0.05 не отвергается")
else:
    print(f"Гипотеза об однородности на уровне значимости 0.05 отвергается")
```

Если будет подозрение на то, что гипотезу о независимости можно не отвергать, то будет вооможность использовать больше критериев для проверки отсутствия эффекта.

Но теперь еще заметим, что на самом деле данные первой и второй выборки согласно гистограмме приблизительно равностепенно раскиданы. А значит, есть смысл рассмотреть гипотезу о равенстве дисперсий, что откроет нам дорогу к двум очень хорошим критериям -- двувыборочному t-тесту (так как данных много и выбросов особо нету) и критерию Манна-Уитни.

Собственно запустим процедуру множественной проверки гипотез с контролем FDR на уровне  0.1 c помощью метода Бенджамини-Иекутелли, где гипотезы будут следующими: гипотеза о независимости, гипотеза о равенстве дисперсий выборок, двувыборочный t-тест и тест Манна-Уитни. Если все гипотезы не отвергнуться, то следовательно и гипотезу об отсутсвии эффекта отклонять не нужно.


```python
from scipy.stats import f
def F_test(x, y):
    x = np.array(x)
    y = np.array(y)
    df1 = len(x) - 1
    df2 = len(y) - 1
    F_stat = np.var(x, ddof=1)/np.var(y, ddof=1)
    pv = 2*np.min([f.cdf(F_stat, df1, df2), 1 - f.cdf(F_stat, df1, df2)])
    return (F_stat, pv)
```

```python
p_vals =  [sps.spearmanr(data[0], data[1])[1], F_test(data[0], data[1])[1],
          sps.ttest_ind(data[0], data[1], equal_var=True)[1], sps.mannwhitneyu(data[0], data[1])[1]]
p_vals
```

```python
multipletests(np.array(p_vals), alpha = al, method = "fdr_by")[0]
```

Таким образом, гипотеза об отсутсвии эффекта не отклоняется.

```python
num_samples = 10000
sizes = [10, 15, 20, 50, 100, 500, 10000]
alphas = [0.01, 0.05, 0.1]
```

```python
def calc_critical_vals(num_samples, sizes, alphas, distr):
    first = distr.rvs((num_samples, sizes[-1]))
    second = distr.rvs((num_samples, sizes[-1]))
    stats_vals = [[] for sz in sizes]
    for x, y in zip(first, second):
        for ind, sz in enumerate(sizes):
            stats_vals[ind].append(sps.ttest_ind(x[:sz], y[:sz])[0])
    result = dict()
    for ind, sz in enumerate(sizes):
        result[sz] = dict()
        for alpha in alphas:
            result[sz][alpha] = (np.round(np.quantile(stats_vals[ind], alpha / 2.0), 3),
                                 np.round(np.quantile(stats_vals[ind], 1.0 - alpha / 2.0), 3))
    return result
```

```python
def write_results(results):
    for sz in sizes:
        for al in alphas:
            print("------------------------------------------------------------------------------------")
            print(f"Критические значения для статистики при n={sz} и a = {al}" +
                 f", полученные моделированием, равны {results[sz][al][0]} и {results[sz][al][1]}")
            print(f"Критические значения распределения Стьюдента -- {sps.t(2*sz - 2).ppf(al / 2)} и {sps.t(2*sz - 2).ppf(1- al/ 2)}")
            print("Критические значения стандартного нормального распределения" + 
                  f" -- {sps.norm().ppf(al / 2)} и {sps.norm().ppf(1- al / 2)}")
            print("------------------------------------------------------------------------------------")
        print("////////////////////////////////////////////////////////////////////////////////////")
        print("////////////////////////////////////////////////////////////////////////////////////")
```

```python
results_t = calc_critical_vals(num_samples, sizes, alphas, sps.t(10))
write_results(results_t)
```

Как можно заметить, для выборок малого размера критические значения лучше приближаются квантилями распределения Стьюдента. Это связано скорее всего с тем, что распределение Стьюдента с 10 степенями свободы уже достаточно сильно похоже на нормальное распределение, а для нормального распределения двустороний t-test -- точный критерий, а не асимптотический!!! 

Проверим, что если мы будем моделировать распредление t-статистики с помощью какого-либо другого распределения, которое меньше похоже на нормальное, то уже будет непонятно, квантили какого распределения лучше описывают критические значения. В качестве примера возьмем стандартное экспоненциальное распределение.

```python
results_exp = calc_critical_vals(num_samples, sizes, alphas, sps.expon())
write_results(results_exp)
```

Как видим, здесь уже непонятно, какие квантили лучше описывают критические значения.

Осталось проверить, что если брать распределение без первых двух моментов, будет все не очень. Рассмотрим распределение Коши.

```python
results_cau = calc_critical_vals(num_samples, sizes, alphas, sps.cauchy())
write_results(results_cau)
```

Как видим, тут уже отклонения от квантилей достаточно большие.

Теперь посчитаем для $\alpha = 0.01, 0.05, 0.1$ реальный уровень значимости, то есть вероятность того, что статистика критерия попадет в критическую область, которая определяется квантилями соотвествующего распределения Cтьюдента. Это еще раз покажет (только  в этот раз проверять не нужно будет глазами), что квантили соответсвующего распределения Стьюдента лучше соотвествуют критическим значениям статистики, нежели квантили нормального распределения. 

```python
def calc_real_significance(num_samples, sizes, alphas, distr_for_model, is_norm_quants):
    first = distr_for_model.rvs((num_samples, sizes[-1]))
    second = distr_for_model.rvs((num_samples, sizes[-1]))
    stats_vals = [[] for sz in sizes]
    for x, y in zip(first, second):
        for ind, sz in enumerate(sizes):
            stats_vals[ind].append(sps.ttest_ind(x[:sz], y[:sz])[0])
    result = dict()
    for ind, sz in enumerate(sizes):
        result[sz] = dict()
        for alpha in alphas:
            low, up = 0, 0
            if is_norm_quants:
                low, up = sps.norm().ppf(alpha / 2), sps.norm().ppf(1 - alpha / 2)
            else:
                low, up = sps.t(2 * sz - 2).ppf(alpha / 2), sps.t(2 * sz - 2).ppf(1 - alpha / 2)
            result[sz][alpha] = float(np.count_nonzero((stats_vals[ind] <= low) | (stats_vals[ind] >= up)))
            result[sz][alpha] = np.round(result[sz][alpha]  / num_samples, 4)
    return result
```

```python
res_st = calc_real_significance(num_samples, sizes, alphas, sps.t(10), False)
```

```python
res_norm = calc_real_significance(num_samples, sizes, alphas, sps.t(10), True)
```

```python
for sz in sizes:
    for al in alphas:
        print("------------------------------------------------------------------------------------")
        print(f"При n={sz} и a = {al} реальный уровень значимости, если брать " +
              f"квантили распределения Стьюдента, равен {res_st[sz][al]}")
        print("При n={sz} и a = {al} реальный уровень значимости, если брать " + 
                 f"квантили нормального распределения, равен {res_norm[sz][al]}")
        print("------------------------------------------------------------------------------------")
        print("////////////////////////////////////////////////////////////////////////////////////")
        print("////////////////////////////////////////////////////////////////////////////////////")
```

**Вывод:** Как видим, на выборках небольшого размера реальный уровень значимости ведет себя лучше если брать квантили распределения Стьюдента. Таким образом, если данные не имеют нормальное распределение, но имеют достаточно похожую к нормальному распредлению природу, стоит пробовать пользоваться двувыборочным t-тестом.

Еще раз подтвердим вывод графиком, который отображает реальный уровень значимости при $\alpha = 0.05$.

```python
def create_plot_real_significance(sizes, num_samples, alpha):  
    first = sps.t(10).rvs((num_samples, sizes[-1]))
    second = sps.t(10).rvs((num_samples, sizes[-1]))
    stats_vals = [[] for sz in sizes]
    for x, y in zip(first, second):
        for ind, sz in enumerate(sizes):
            stats_vals[ind].append(sps.ttest_ind(x[:sz], y[:sz])[0])
    result_n = dict()
    result_t = dict()
    for ind, sz in enumerate(sizes):
        result_n[sz] = dict()
        result_t[sz] = dict()
        for alpha in alphas:
            low_n, up_n = sps.norm().ppf(alpha / 2), sps.norm().ppf(1 - alpha / 2)
            low_t, up_t = sps.t(2 * sz - 2).ppf(alpha / 2), sps.t(2 * sz - 2).ppf(1 - alpha / 2)
            result_n[sz][alpha] = float(np.count_nonzero((stats_vals[ind] <= low_n) 
                                                         | (stats_vals[ind] >= up_n)))
            result_n[sz][alpha] = np.round(result_n[sz][alpha]  / num_samples, 4)
            result_t[sz][alpha] = float(np.count_nonzero((stats_vals[ind] <= low_t) 
                                                         | (stats_vals[ind] >= up_t)))
            result_t[sz][alpha] = np.round(result_t[sz][alpha]  / num_samples, 4)
    
    title_norm = "Реальный уровень значимости для нормального распределения"
    title_t = "Реальный уровень значимости для распределении Cтьюдента"
    reals_norm, reals_t = [], []
    for v_n, v_t in zip(result_n.values(), result_t.values()):
        reals_norm.append(v_n[0.05])
        reals_t.append(v_t[0.05])
    plt.figure(figsize = (16, 9))
    plt.plot(sizes, reals_norm, linewidth = 3, color = 'red', label = 'Нормальное распределение')
    plt.plot(sizes, reals_t, linewidth = 3, color = 'blue', label = 'Распределение Стьюдента')
    plt.xlabel("Размер выборки")
    plt.ylabel("Реальный уровень значимости")
    plt.legend()
    plt.title(fr"График реальных уровней значимости при $\alpha = {alpha}$")
    plt.show()
```

```python
create_plot_real_significance(np.arange(10, 101, 5), 2000, 0.05)
```

Еще раз видим, что реальный уровень оказазался лучше у Стьюдента.
