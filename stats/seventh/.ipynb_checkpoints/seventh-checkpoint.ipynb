```python
sps.kruskal(*[corr_frame[corr_frame['drug'] == drug]['time'].values for drug in drugs])
```

Отлично, кажется эффект дейсвительно есть. Для нахождения лекарства, которое имеет эффект, воспользуемся методами Шеффе и LSD Фишера. Так как дисперсии у нас не равны, то критерий Шеффе в теории ведет себя лучше, а также он удобнее для нас, так как мы знаем что есть плацебо, но проверим оба критерия. Реализуем их сами (используя код семинара). В качестве приблизительного pvalue в LSD-Фишера будем брать модуль разности / на коэффициент, который умножается на квантиль распределения Стьюдента. Лучше конечно воспользоваться моделированием, но для приближения подойдет.

```python
##  Lsd-Fisher

dr_names_sorted = corr_frame.groupby('drug')['time'].mean().sort_values().index
ssamples = [corr_frame[corr_frame['drug'] == drug]['time'].values for drug in dr_names_sorted]
alpha = 0.1
p_vals_lsd = []
for j in range(len(drugs) - 1):
    nj = len(ssamples[j])
    njj = len(ssamples[j + 1])
    coef = np.sqrt(
        (nj + njj) / (nj * njj)
    ) * np.sqrt(
        (
            (nj - 1) * ssamples[j].var(ddof=False) + (njj - 1) * ssamples[j + 1].var(ddof=False)
        ) / (nj + njj - 2)
    )
    LSD = sps.t.ppf(
        1 - alpha, 
        nj + njj - 2
    ) * coef
    approx_pval = 1 - sps.t(nj + njj - 2).cdf(abs(ssamples[j].mean() - ssamples[j + 1].mean()) / coef)
    p_vals_lsd.append(approx_pval)
    print("{} with {}: Diff = {}, LSD = {}, approx. pval = {}".format(dr_names_sorted[j],
                                                                      dr_names_sorted[j + 1],
                                                                      abs(ssamples[j].mean()
                                                                          - ssamples[j + 1].mean()),
                                                                      LSD, approx_pval))
```

Нужно не забыть сделать множественную проверку гипотез для LSD!

```python
p_vals = p_vals_lsd + p_vals_norm
ans = multipletests(np.array(p_vals), alpha =alpha, method = "holm")[0]
ans
```

Как видим, отверглась гипотеза о равенстве эффектов от Codeine и Irunine (который согласно условию вроде как плацебо). Значит не отвергаем гипотезу о том, что есть факт от CodeIne!

```python
print(dr_names_sorted)
```

Запомним, что под индексом один у нас плацебо, чтобы правильно заполнить столбец c.

```python
def sheffe(c, samples):
    ns = np.array([len(x) for x in samples])
    k = len(samples)
    n = np.array(ns).sum()
    sin = 1. / (n - k) * (ns * np.array([x.var() for x in samples])).sum()
    S = (c * np.array([x.mean() for x in samples])).sum() ** 2 / ((k - 1) * sin * (c ** 2 / ns).sum())
    return (S, 1 - sps.f( len(dr_names_sorted) - 1, len(frame["drug"].to_list()) - 1).cdf(S))

u = sps.f.ppf(1-0.1, len(dr_names_sorted) - 1, len(frame["drug"].to_list()) - 1)

print("Квантиль распределения Фишера: {}\n".format(u))

for i in range(6):
    if i == 1:
        continue
    c = np.zeros(6)
    c[1] = 1
    c[i] = -1
    st = sheffe(c, ssamples)
    print("sheffe for {} and {} is {}, pvalue -- {}".format(dr_names_sorted[i], 
                                                  dr_names_sorted[1],st[0], st[1]))
```

А тут оказывается, что эффекта нету ( множественную проверку запускать не надо, все pvalue большие). Это может быть связано с тем, что критерий Шеффе более консервативен. С учетом того, что наличие эффекта было показано методом краскела-Уоллиса, верим LSD.

Тем не менее, проверим нормальность выборок и равенство дисперсий, как и хотели ранее.

```python
p_vals = []
for fir in range(num_first):
    for sec in range(num_second):
        p_vals.append(sps.shapiro(frame[(frame['first'] == fir) & (frame['second'] == sec)]['val'])[1])
p_vals    
```

```python
ans = multipletests(np.array(p_vals), alpha = 0.05, method = "holm")[0]
ans
```

К сожалению, нормальность некоторых выборок отвергнута. но всего двух. Проверим равенство дисперсий.

```python
sps.levene(*[frame[(frame['first'] == f) & (frame['second'] == s)]['val'].values 
                           for f in range(5) for s in range(5)], center='mean')[1]
```

Гипотеза о равестве дисперсий не отвергается, но для пары выборок гипотезу о нормальности отвергает. Тем не менее, в совокуности с тем, что длины выборок равны, считаем что критерий Фишера применять можно. Реализуем его, затем сравним с библиотекой.

```python
def fisher_test2(frame):

    first_vals = frame["first"].unique()
    second_vals = frame["second"].unique()
    k, m =  len(first_vals), len(second_vals)
    n = len(frame) / k / m
    vals = frame['val'].values
    al = frame['val'].values.mean()
    fir = frame.groupby(by=['first']).mean()['val'].values
    sec = frame.groupby(by=['second']).mean()['val'].values
    fir_sec =  frame.groupby(by=['first', 'second']).mean().values
    
    s_1 = 10 * m  * np.sum(np.power(fir - al, 2)) / (k - 1)
    s_2 = 10 * k *  np.sum(np.power(sec - al, 2)) / (m - 1)
    s_12 = 10 * np.sum(np.power(fir_sec.reshape((k, m)) - fir.reshape((k, 1)) - sec.reshape((1, m)) + al , 2)
                     ) / ((m - 1) * (k - 1))
    s_int = np.sum(np.sum(np.power(vals.reshape(k , m, 10) - fir_sec.reshape(k , m, 1), 2), axis = (1, 2))
                  ) / (k * m * (n - 1))
    f1 = s_1 / s_int
    f2 = s_2 / s_int
    f12 = s_12 / s_int
    res = []
    res.append(['H1', f1, 1 - sps.f( k - 1, k * m * (n - 1)).cdf(f1)])
    res.append(['H2', f2, 1 - sps.f( m - 1, k * m * (n - 1)).cdf(f2)])
    res.append(['H3', f12, 1 - sps.f( (k - 1)*(m - 1), k * m * (n - 1)).cdf(f12)])
    return res
```

```python
res = fisher_test2(frame)
res_frame = pd.DataFrame(data = res, columns = ["hypo", "stat value", "pvalue"])
res_frame
```

Как видим, только гипотеза об отсутсвии межфакторного взаимодействия не отвергается. Сверимся с библиотекой statsmodels.stats.anova.

```python
from statsmodels.stats.anova import anova_lm
from statsmodels.formula.api import ols
```

```python
model = ols('val ~ C(first) + C(second) + C(first):C(second)', data=frame).fit()
anova_lm(model)
```

### Задача 5

Много переменных нужно проварьировать, поэтому будем их варировать по очереди. Начнем с функции распределения F и количества объектов n. Возьмем пока $\beta = 0.15$,$ k = 7$, $\mu = 7$ . Альфы будут выбираться равномерно из отрезка от 0 до 10. В качестве функций ошибок рассмотрим стьюдента с 10 степенями свободы, нормальное и равномерное распределение на отрезках [-2,2].

```python
mu = 7
junk = [sps.norm(), sps.t(10), sps.uniform(-2, 2)]
titles_junk = ['normal', 't', 'U(-2, 2)']
distr_alpha = sps.uniform(0, 20)
n_s = np.append(np.array([5, 7, 10]),np.arange(15, 51, 5))
k = 7
beta = 0.15
```

```python
def model_result(n, k, beta, mu, err, distr_alpha, sig = 0.05, num_tries = 200):
    page_rej, fridman_rej = 0, 0 
    for _ in range(num_tries):
        alphas = distr_alpha.rvs(n)
        errors = err.rvs(size = (n, k))
        data = np.ones((n, k))
        for i in range(n):
            for j in range(1, k + 1):
                data[i][j - 1] = mu + alphas[i] + j * beta + errors[i][j - 1]
        test_p =  sps.page_trend_test(data)
        test_f = sps.mstats.friedmanchisquare(*[data[:, j] for j in range(k)])
        #print(test_p, test_f)
        if sps.page_trend_test(data).pvalue < sig:
            page_rej += 1
        if sps.mstats.friedmanchisquare(*[data[:, j] for j in range(k)])[1] < sig:
            fridman_rej += 1
    return page_rej / num_tries, fridman_rej / num_tries
```

```python
plt.figure(figsize = (16, 9))
result_f = [[] for _ in junk]
result_p = [[] for _ in junk]
for ind, err in enumerate(junk):
    for n in n_s:
        res_p, res_f = model_result(n, k , beta, mu, err, distr_alpha)
        result_f[ind].append(res_f)
        result_p[ind].append(res_p)
for ind, res_f in enumerate(result_f):
    plt.plot( n_s, res_f, label = titles_junk[ind] + ' Friedman', linewidth = 4)
for ind, res_p in enumerate(result_p):
    plt.plot( n_s, res_p, label = titles_junk[ind] + ' Page', linewidth = 4)
plt.legend()
plt.xlabel("Количество элементов")
plt.ylabel("Мощность критерия, полученная моделированием")
plt.show()
```

Как видим, при одинаковых ошибках действительно критерий Пейджа мощнее. Также, с ростом n (числа элементов), мощности каждого из критериев растут независимо от ошибки. Также видно, что мощность критериев также зависит от функции распределения ошибки. Например, если ошибка имеет равномерное распределение, каждый из критериев ведет себя стабильно мощнее, чем если ошибка взята из нормального распредления. 

Теперь поварьируем k и n, ошибку возьмем нормальную, $\beta = 0.15$

```python
k_s = np.arange(3, 12, 2)
plt.figure(figsize = (20, 11))
result_f = [[] for _ in k_s]
result_p = [[] for _ in k_s]
for ind, k in enumerate(k_s):
    for n in n_s:
        res_p, res_f = model_result(n, k, beta, mu, sps.norm(), distr_alpha)
        result_f[ind].append(res_f)
        result_p[ind].append(res_p)
for ind, res_f in enumerate(result_f):
    plt.plot( n_s, res_f, label =  f' k={k_s[ind]}' + ' Fridman', linewidth = 4)
for ind, res_p in enumerate(result_p):
    plt.plot( n_s, res_p, label = f' k={k_s[ind]}' + ' Page', linewidth = 4)
plt.legend()
plt.xlabel("Количество элементов")
plt.ylabel("Мощность критерия, полученная моделированием")
plt.show()
```

Как видим, чем больше k, тем мощнее критерии. Отлично, осталось поварьировать $\beta$.

```python
beta_s = [0.02, 0.05, 0.1, 0.15, 0.2, 0.3]
n_s = np.append(np.array([5, 7, 10]),np.arange(15, 31, 5))
```

```python
plt.figure(figsize = (22, 12))
result_f = [[] for _ in beta_s]
result_p = [[] for _ in beta_s]
for ind, beta in enumerate(beta_s):
    for n in n_s:
        res_p, res_f = model_result(n, k, beta, mu, sps.norm(), distr_alpha)
        result_f[ind].append(res_f)
        result_p[ind].append(res_p)
for ind, res_f in enumerate(result_f):
    plt.plot( n_s, res_f, label =  f' b={beta_s[ind]}' + ' Fridman', linewidth = 5)
for ind, res_p in enumerate(result_p):
    plt.plot( n_s, res_p, label = f' b={beta_s[ind]}' + ' Page', linewidth = 5)
plt.legend()
plt.xlabel("Количество элементов")
plt.ylabel("Мощность критерия, полученная моделированием")
plt.show()
```

Как видим, при увеличении $\beta$ мощность увеличивается, что логично.

 **Вывод**: Как видим, при увеличении любого из параметров n, k и $\beta$  мощность критериев увеличивается. Также видно, что на маленьких значениях количества наблюдений критерий Пэйджа сильно мощнее критерия Фридмана, поэтому в случае альтернативы на неравенства использовать нужно именно его!
