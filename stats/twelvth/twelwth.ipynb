```python
import statsmodels.tsa.api as smt
import statsmodels
```

```python
def tsplot(y, lags=None, figsize=(12, 7), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))

        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.5)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.5)

        print("Критерий Дики-Фуллера: p=%f" % statsmodels.tsa.stattools.adfuller(y)[1])

        plt.tight_layout()
    return 
```

```python
tsplot(data_x[0], lags=30)
```

Тест Дики-Фуллера говорит нам о том, что гипотеза о нестационарности ряда отвергается. Больших лагов вроде нету. Однако, попробуем еще 1 раз продифферентцировать ряд, и посмотреть графики частичной и авто корреляции.

```python
data_diff = data_x[0].diff(1)
tsplot(data_diff[1:], lags=30)
```

Как видим, при дифференцировании график частичной корреляции сильно упал, а значит дифференцировать не нужно. Таким образом, есть предположение о том, что ряд скорее всего есть белый шум. Если получим, что лучшие параметры для ARIMA -- нули, то тем самым получим, что лучше всего предсказывает 

```python
ps = range(0, 10)
d = [0, 1]
qs = range(0, 10)

from itertools import product
from tqdm import tqdm

parameters = product(ps,d,  qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
results = []
best_aic = float("inf")

for param in tqdm(parameters_list):
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model = statsmodels.tsa.statespace.sarimax.SARIMAX(data_x[0], 
                                                           order=(param[0],  param[1], param[2])).fit(disp=-1)

#         model=sm.tsa.ARIMA(data_boxcox.dropna(), order=(param[0], d, param[1]), ).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except Exception:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])

warnings.filterwarnings('default')

result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
```

```python
result_table.sort_values(by = 'aic', ascending=True).head(10)
```

Как видим, действительно лучшие параметры -- нули.

Визуализируем результат для сверки.

Предположим, что период 120. Побробуем убрать сезонность.

Теперь уберем сезонность с периодом 12.

```python
data_without_first_seasonal = data_train['y'] - result.seasonal
result_second = STL(data_without_first_seasonal, period=12).fit()
result_second.plot();
```

Не хочень похоже на то, что мы убрали всю сезонность, однако проверим тестом Дики-Фуллера и KPSS на однородность.

```python
fuller_resid = adfuller(result_second.resid, regression='c')[1]
fuller_resid
```

```python
kpss_resid = kpss(result_second.resid, regression='c')[1]
kpss_resid
```

```python
pvals = np.array([fuller_resid, kpss_resid])
fuller_resid_adj, kpss_resid_adj = multipletests(pvals=pvals, method='holm')[1]
fuller_resid_adj, kpss_resid_adj
```

Тесты говорят, что остатки стационарные (тест Дики-Фуллера отвергвет нестационарность, КПСС-тест не отвергает стационарность), поэтому будем считать, что остатки стационарны. Также с помощью модели ARIMA будем учитывать сезонность 12. Посмотрим еще на коррелограммы полученных остатков.
