```python
# Useful Constants

# Those are separate normalised input features for the neural network
INPUT_SIGNAL_TYPES = [
    "body_acc_x_",
    "body_acc_y_",
    "body_acc_z_",
    "body_gyro_x_",
    "body_gyro_y_",
    "body_gyro_z_",
    "total_acc_x_",
    "total_acc_y_",
    "total_acc_z_"
]

# Output classes to learn how to classify
LABELS = [
    "WALKING", 
    "WALKING_UPSTAIRS", 
    "WALKING_DOWNSTAIRS", 
    "SITTING", 
    "STANDING", 
    "LAYING"
]

DATA_PATH = "./"

DATASET_PATH = DATA_PATH + "UCI HAR Dataset/"
print("\n" + "Dataset is now located at: " + DATASET_PATH)

TRAIN = "train/"
TEST = "test/"


# Load "X" (the neural network's training and testing inputs)

def load_X(X_signals_paths):
    X_signals = []
    
    for signal_type_path in X_signals_paths:
        file = open(signal_type_path, 'r')
        # Read dataset from disk, dealing with text files' syntax
        X_signals.append(
            [np.array(serie, dtype=np.float32) for serie in [
                row.replace('  ', ' ').strip().split(' ') for row in file
            ]]
        )
        file.close()
    
    return np.transpose(np.array(X_signals), (1, 2, 0))

X_train_signals_paths = [
    os.path.join(*[DATASET_PATH, TRAIN, "Inertial Signals/", signal+"train.txt"]) for signal in INPUT_SIGNAL_TYPES
]
X_test_signals_paths = [
    os.path.join(*[DATASET_PATH, TEST, "Inertial Signals/", signal+"test.txt"]) for signal in INPUT_SIGNAL_TYPES
]

X_train = load_X(X_train_signals_paths)
X_test = load_X(X_test_signals_paths)


# Load "y" (the neural network's training and testing outputs)

def load_y(y_path):
    file = open(y_path, 'r')
    # Read dataset from disk, dealing with text file's syntax
    y_ = np.array(
        [elem for elem in [
            row.replace('  ', ' ').strip().split(' ') for row in file
        ]], 
        dtype=np.int32
    )
    file.close()
    
    # Substract 1 to each output class for friendly 0-based indexing 
    return y_ - 1

y_train_path = os.path.join(DATASET_PATH, TRAIN, "y_train.txt")
y_test_path = os.path.join(DATASET_PATH, TEST, "y_test.txt")

y_train = load_y(y_train_path)
y_test = load_y(y_test_path)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pytorch_lightning as pl
from pytorch_lightning.callbacks import ModelCheckpoint, EarlyStopping
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

```python
class ConvBlock(nn.Module):
    def __init__(self, in_channels: int, out_channels: int, kernel_size=3):
        super().__init__()

        self.conv = nn.Conv1d(in_channels, out_channels,
                              kernel_size=kernel_size)
        self.act = nn.Mish()
        self.norm = nn.BatchNorm1d(out_channels)

    def forward(self, x: torch.tensor):
        return self.act(self.norm(self.conv(x)))
```

```python
def test_conv_block():
    batch_size = 3
    seq_len = 128
    channels = 9
    x = torch.randn((batch_size, channels, seq_len))
    block = ConvBlock(channels, 16)
    assert block(x).shape == (batch_size, 16, seq_len - 2)
    print("OK!")

test_conv_block()
```

```python
class LinearBlock(nn.Module):
    def __init__(self, in_channels: int, out_channels: int):
        super().__init__()

        self.linear = nn.Linear(in_channels, out_channels)
        self.act = nn.LeakyReLU()
        self.norm = nn.BatchNorm1d(out_channels)

    def forward(self, x: torch.tensor):
        return self.act(self.norm(self.linear(x)))
```

```python
def test_linear_block():
    batch_size = 3
    channels = 9
    x = torch.randn((batch_size, channels))
    block = LinearBlock(channels, 16)
    assert block(x).shape == (batch_size, 16)
    print("OK!")

test_linear_block()
```

```python
class HARClassifier(nn.Module):

    CONV_KERNELS = (3, 2, 2, 2)

    @staticmethod
    def init_weights(m):
        if isinstance(m, nn.Linear):
            torch.nn.init.xavier_uniform_(m.weight)
            m.bias.data.fill_(0.01)

    def __init__(self, seq_len: int, n_channels: int, num_classes: int):
        super().__init__()

        self.conv_features_extractor = nn.Sequential(
            ConvBlock(n_channels, 16, kernel_size=3),
            nn.MaxPool1d(kernel_size=2),
            ConvBlock(16, 32, kernel_size=2),
            nn.MaxPool1d(kernel_size=2),
            ConvBlock(32, 64, kernel_size=2),
            nn.MaxPool1d(kernel_size=2),
            ConvBlock(64, 128, kernel_size=2),
            nn.MaxPool1d(kernel_size=2),
            nn.Flatten(),
        )

        len_after = seq_len
        for kernel_size in self.CONV_KERNELS:
            len_after -= kernel_size - 1
            len_after //= 2

        self.classifier = nn.Sequential(
            LinearBlock(len_after * 128, 256),
            LinearBlock(256, 128),
            LinearBlock(128, 64),
            nn.Dropout(p=0.3),
            nn.Linear(64, num_classes)
        )

        self.apply(self.init_weights)

    def forward(self, x: torch.tensor):
        features_conv = self.conv_features_extractor(x)
        return self.classifier(features_conv)
```

```python
def test_classifier():
    batch_size = 3
    seq_len = 128
    channels = 9
    n_classes = 9
    x = torch.randn((batch_size, channels, seq_len))
    model = HARClassifier(seq_len, channels, n_classes)
    assert model(x).shape == (batch_size, n_classes)
    print("OK!")

test_classifier()
```

```python
def accuracy(y_pred, y_true):
    return torch.mean((torch.argmax(y_pred, dim=-1) == y_true).float())
```

```python
def test_accuracy():
    y_pred = torch.tensor([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
    y_true = torch.tensor([0, 1, 0])
    assert accuracy(y_pred, y_true) == 2 / 3
    print("OK!")

test_accuracy()
```

```python
class HARClassifierWrapper(pl.LightningModule):
    def __init__(self, seq_len: int, n_channels: int, num_classes: int):
        super().__init__()

        self.model = HARClassifier(seq_len, n_channels, num_classes)
        self.criterion = nn.CrossEntropyLoss()

    def forward(self, x: torch.tensor):
        return self.model(x)

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=1e-2)
        return optimizer

    def training_step(self, train_batch, batch_idx):
        x, y = train_batch
        y_pred = self.model(x)
        loss = self.criterion(y_pred, y)
        self.log("train_loss", loss)
        self.log("train_accuracy", accuracy(y_pred, y), prog_bar=True)
        return loss

    def validation_step(self, val_batch, batch_idx):
        x, y = val_batch
        y_pred = self.model(x)
        loss = self.criterion(y_pred, y)
        self.log("val_loss", loss, prog_bar=True)
        self.log("val_accuracy", accuracy(y_pred, y), prog_bar=True)
```

```python
class HARDataset(Dataset):
    def __init__(self, X: np.ndarray, y: np.ndarray):
        super().__init__()

        self.X = torch.from_numpy(X.reshape(len(X), X.shape[2], -1))
        self.y = torch.from_numpy(y).squeeze().long()

    def __getitem__(self, idx: int):
        return self.X[idx], self.y[idx]

    def __len__(self):
        return len(self.X)
```

```python
train_dataset = HARDataset(X_train, y_train)
val_dataset = HARDataset(X_test, y_test)
```

```python
batch_size = 512
_, seq_len, channels = X_train.shape
num_classes = len(np.unique(y_train))
```

```python
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
```

```python
model = HARClassifierWrapper(seq_len, channels, num_classes).to(device)
```

```python
checkpoint_callback = ModelCheckpoint(
    save_top_k=1,
    monitor="val_accuracy",
    mode="max",
    dirpath="../checkpoints",
    filename="HARClassifier-{epoch:02d}-{val_accuracy:.2f}",
)
```

```python
early_stop_callback = EarlyStopping(
    monitor="val_accuracy",
    min_delta=0.0,
    patience=40,
    verbose=False,
    mode="max"
)
```

```python
trainer = pl.Trainer(
    gpus=1,
    log_every_n_steps=2,
    gradient_clip_val=10.0,
    max_epochs=200,
    callbacks=[
        checkpoint_callback,
        early_stop_callback
    ]
)
```

```python
trainer.fit(model, train_loader, val_loader)
```

```python
checkpoint_callback.best_model_path
```

```python
model_validation = HARClassifierWrapper.load_from_checkpoint(
    checkpoint_callback.best_model_path,
    seq_len=seq_len,
    n_channels=channels,
    num_classes=num_classes
).to(device)
```

```python
model_validation.eval()
acc_value = 0
for X, y in val_loader:
    X = X.to(device)
    y = y.to(device)
    y_pred = model_validation(X.to(device))
    acc_value += accuracy(y_pred, y).item() * len(y)

acc_value /= len(val_loader.dataset)
```

```python
SIZE = 224
BATCH_SIZE = 32
```

```python
class DogsDataset(Dataset):
    def __init__(self, 
                 image_paths: np.ndarray, 
                 labels: np.ndarray = None,
                 mode: str = "train", 
                 device: torch.device = torch.device("cpu")):
        super().__init__()

        if mode not in ["train", "test"]:
            raise ValueError(f"Incorrect mode: {mode}")

        self.mode = mode
        self.image_paths = image_paths
        self.labels = labels
        self.cache = dict()
        self.to_tensor_transform = transforms.ToTensor()

    def load_sample(self, file: str) -> np.array:
        if file in self.cache:
            return self.cache[file]
        image = cv2.imread(file)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        self.cache[file] = cv2.resize(image, (SIZE, SIZE))
        return self.cache[file]

    def __getitem__(self, idx: int):
        x = self.load_sample(self.image_paths[idx])
        if self.mode == "test":
            return x
        return x, self.labels[idx]

    def __len__(self):
        return len(self.image_paths)
```

```python
dogs_dataset = DogsDataset(train_files, train_labels)
```

```python
test_ratio = 0.1
test_size = int(len(dogs_dataset) * test_ratio)
train_size = len(dogs_dataset) - test_size
data_train, data_val = random_split(dogs_dataset, (train_size, test_size))
```

```python
train_loader = DataLoader(data_train, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(data_val, batch_size=BATCH_SIZE)
```

### Моделька

```python
def accuracy(y_pred: torch.tensor, y_true: torch.tensor) -> float:
    return torch.mean((torch.argmax(y_pred, dim=-1) == y_true).float())
```

Возьмём за основу EfficientNet и зафайнтуним её

```python
class LinearBlock(nn.Module):
    def __init__(self, in_channels: int, out_channels: int):
        super().__init__()

        self.linear = nn.Linear(in_channels, out_channels)
        self.act = nn.LeakyReLU()
        self.norm = nn.BatchNorm1d(out_channels)

    def forward(self, x: torch.tensor):
        return self.act(self.norm(self.linear(x)))
```

```python
class DogsClassifier(pl.LightningModule):
    def __init__(self, n_classes: int):
        super().__init__()
        self.train_transforms = iaa.RandAugment(n=(0, 10), m=(5, 25))
        self.valid_transforms = iaa.Identity()
        self.norm = transforms.Normalize([0.485, 0.456, 0.406],
                                         [0.229, 0.224, 0.225])

        self.backbone = efficientnet_b3(pretrained=True)
        self.backbone.classifier = nn.Identity()

        self.predictor = nn.Sequential(
            nn.Dropout(p=0.3, inplace=True),
            nn.Linear(1536, n_classes)
        )

        self.criterion = nn.CrossEntropyLoss()

    def transform_input(self, x: torch.tensor):
        if self.training:
            x = self.train_transforms(images=x.cpu().detach().numpy())
        else:
            x = self.valid_transforms(images=x.cpu().detach().numpy())

        x = np.transpose(x, (0, 3, 1, 2))
        x = self.norm(torch.from_numpy(x / 255).float()).to(device)
        return x

    def forward(self, x: torch.tensor):
        x = self.transform_input(x)
        return self.predictor(self.backbone(x))

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(
            filter(lambda p: p.requires_grad, self.parameters()), lr=1e-3)
        return optimizer

    def training_step(self, train_batch, batch_idx):
        x, y = train_batch
        y_pred = self(x)

        loss = self.criterion(y_pred, y)
        self.log("train_loss", loss)
        self.log("train_accuracy", accuracy(y_pred, y), prog_bar=True)

        return loss

    def validation_step(self, val_batch, batch_idx):
        x, y = val_batch
        y_pred = self(x)

        loss = self.criterion(y_pred, y)
        self.log("val_loss", loss, prog_bar=True)
        self.log("val_accuracy", accuracy(y_pred, y), prog_bar=True)
```

```python
checkpoint_callback = ModelCheckpoint(
    save_top_k=1,
    monitor="val_accuracy",
    mode="max",
    dirpath="../checkpoints",
    filename="DogsClassifier-{epoch:02d}-{val_accuracy:.2f}",
)
```

```python
early_stop_callback = EarlyStopping(
    monitor="val_accuracy",
    min_delta=0.0,
    patience=3,
    verbose=False,
    mode="max"
)
```

```python
backbone_finetuning = BackboneFinetuning(1)
```

```python
trainer = pl.Trainer(
    gpus=1 if torch.cuda.is_available() else 0,
    log_every_n_steps=2,
    gradient_clip_val=10.0,
    max_epochs=5,
    callbacks=[
        checkpoint_callback,
        early_stop_callback,
        backbone_finetuning,
    ]
)
```

```python
model = DogsClassifier(n_classes).to(device)
```

```python
trainer.fit(model, train_loader, val_loader)
```
