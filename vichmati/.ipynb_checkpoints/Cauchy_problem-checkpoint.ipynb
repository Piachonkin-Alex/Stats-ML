```python
import numpy as np
import matplotlib.pyplot as plt
```

В данной программе нужно написат решение задачи Коши методом Рунге-Кутты с помощью следующей таблицы Бутчера:

Проверим условия сходимости, чтобы определеить порядок точности:
$$b_1 + b_2 = 1$$
$$b_1c_1 + b_2c_2 = \frac{1}{2}$$
$$b_1c_1^2 + b_2c_2^2 = \frac{10}{32} \neq \frac{1}{3}$$
Значит, имеем второй порядок точности.

Реализуем метод Ньютона:

```python
def Newton(x0, f, gradf, max_iter=250, tol=1e-5):
    x = x0
    
    for k in range(max_iter):
        if np.isscalar(gradf(x)):
            dif = f(x) / gradf(x)
            if np.abs(dif) < tol:
                break
        else:
            dif = np.linalg.inv(gradf(x)) @ f(x)
            if np.linalg.norm(dif, ord=np.inf) < tol:
                break
        
        x -= dif
    
    return x
```

Теперь реализуем метод Рунге-Кутты для решения задачи Коши:

```python
def make_arange(segment, n):
    h = segment[1] / float(n)
    t = np.arange(n+1) * h
    return t

def Cauchy(segment, u0, f, gradf, n, tol=1e-5):
    u = np.empty(n+1, dtype=object)
    u[0] = u0
    h = segment[1] / float(n)
    t = make_arange(segment, n)
    
    k1 = None
    k2 = None
    
    def gen_f_k1(x, i):
        return x - f(t[i] + h / 4, u[i] + x * h / 4  )
    
    def gen_gradf_k1(x, i):
        if np.isscalar(x):
            I = 1
        else:
            I = np.eye(x.shape[0])
        return I - h/4 * gradf(t[i] + h / 4, u[i] + x * h / 4)
    
    def gen_f_k2(x, i):
        return x - f(t[i] + 3 * h / 4, u[i] + h * (2 * k1 + x) / 4)
    
    def gen_gradf_k2(x, i):
        I = None
        if np.isscalar(x):
            I = 1
        else:
            I = np.eye(x.shape[0])
        return I - h * gradf(t[i] + 3 * h / 4, u[i] + h * (2 * k1+x) / 4) / 4
    
    for i in range(n):
        f_k1 = lambda x: gen_f_k1(x, i)
        gradf_k1 = lambda x: gen_gradf_k1(x, i)
        k1 = Newton(0, f_k1, gradf_k1, tol=tol)
        
        f_k2 = lambda x: gen_f_k2(x, i)
        gradf_k2 = lambda x: gen_gradf_k2(x, i)
        k2 = Newton(k1 * h / 2, f_k2, gradf_k2, tol=tol)
        
        u[i+1] = u[i] + h/2 * (k1 + k2)
        
    return u
        
```

Напишем функцию для графика, на котором изображено точное и численное решение

```python
def plot(segment, my_sol, real_sol):
    t = make_arange(segment, my_sol.shape[0]-1)
    t1 = make_arange(segment, 1000)
    plt.figure(figsize=(16, 9))
    plt.scatter(t, my_sol, label="численное решение")
    plt.plot(t1, real_sol(t1), label="точное решение", color='g')
    plt.legend()
    plt.show()
```

Проверим решение на  $u_t = e^t$,  $u(0) = 0$

```python
def f(t, u):
    return np.exp(t)

def gradf(t, u):
    return 0

segment = [0, 1]

my_sol = Cauchy(segment, 1, f, gradf, 10)
```

```python
plot(segment, my_sol, np.exp)
```

$u_t = v$
$v_t = -u$
$u(0) = 0, v(0) = 1$

```python
segment2 = [0, np.pi/2]

def f2(t, u):
    return np.array([u[1], -u[0]])

def gradf2(t, u):
    return np.array([[0,1], [-1,0]])

my_sol2 = Cauchy(segment2, np.array([0.,1.]), f2, gradf2, 10)
```

```python
my_sin = np.array([my_sol2[i][0] for i in range(my_sol2.shape[0])])
my_cos = np.array([my_sol2[i][1] for i in range(my_sol2.shape[0])])
```

```python
plot(segment2, my_sin, np.sin)
```

```python
plot(segment2, my_cos, np.cos)
```

Построим график зависимости ошибки от размера сетки

```python
def plot_error(segment, u0, f, gradf, real_sol):
    ns = np.arange(1, 251)
    dim = len(real_sol)
    errors = np.array([])
    
    for n in ns:
        sol = Cauchy(segment, u0, f, gradf, n, tol=1e-10)
        t = make_arange(segment,n)
        if dim == 1:
            errors = np.append(errors, np.max(np.abs(sol - real_sol[0](t))))
        else:
            diff = np.zeros(n)
            for ind in range(dim):
                err = real_sol[ind](t) - sol[ : , ind]
                diff = diff + err * err
            errors = np.append(errors,errors.append(np.max(diff)))    
 
    plt.figure(figsize=(16, 9))
    plt.semilogy(ns, errors)
    plt.xlabel("Количество узлов сетки")
    plt.title("Численное решение")
    plt.show()
```

```python
plot_error(segment, 1, f, gradf, [np.exp])
```

А также график ошибки в лог. шкале:

```python
def plot_log_error(segment, u0, f, gradf, real_sol):
    ns = 2**(np.arange(1, 8))
    dim = len(real_sol)
    errors = np.array([])
    
    for n in ns:
        sol = Cauchy(segment, u0, f, gradf, n, tol=1e-10)
        t = make_arange(segment,n)
        if dim == 1:
            errors = np.append(errors, np.max(np.power(sol - real_sol[0](t), 2)))
        else:
            diff = np.zeros(n)
            for ind in range(dim):
                err = real_sol[ind](t) - sol[ : , ind]
                diff = diff + err * err
            errors = np.append(errors,errors.append(np.max(np.sqrt(diff))))    
 
    plt.figure(figsize=(14, 10))
    plt.plot(np.log(ns), errors)
    plt.xlabel("Количество узлов сетки")
    plt.title("Численное решение")
    plt.yscale('log')
    plt.show()
```

```python
plot_log_error(segment, 1, f, gradf, [np.exp])
```

По наклону графика ошибки видим, что на одно деление вдоль оси x приходится примерно два деления вдоль оси y, т. е. порядок аппроксимации второй. Это согласуется с теоретическими результатами.
